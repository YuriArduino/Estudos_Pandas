`````python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/YuriArduino/Estudos_Pandas/refs/heads/data-tests/aluguel.csv'
dados = pd.read_csv(url, sep=';')
dados.head()
`````

#Criando colunas numéricas

`````python
#Cria uma nova coluna usando duas existentes
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
`````

`````python
# Preenche os valores ausentes (NaN) com 0 e modifica o DataFrame no lugar
dados.fillna(0,inplace = True)
dados['Valor_por_mes'] = dados['Valor'] + dados['Condominio']
dados.head()
`````

`````python
dados['Valor_por_ano'] = dados['Valor_por_mes'] * 12 + dados['IPTU']
dados.head()
`````

#Criando colunas Categóricas

`````python
#Por ser string vamos concaternar as informações
dados['Descricao'] = dados['Tipo'] + dados['Bairro']
dados.head()
`````

`````python
#Melhorando a visualização
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro']
dados.head()
`````

## Convertendo dados

`````python
dados.dtypes
`````

É necessário `converter` informações (int) para string usando o método .asType() e informando para qual tipo será feita a conversão.

`````python
dados['Descricao'] = dados['Tipo'] + ' em ' + dados['Bairro'] + ' com ' + \
                                        dados['Quartos'].astype(str) + ' quarto(s) ' + \
                                        ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
dados.head()


# \ = pula linha
#Podemos apenas concatenar str + str
`````

#Criando uma coluna binária

Método .`apply()`, utilizado para aplicar uma determinada função em um conjunto específico de dados. No nosso caso, esse conjunto é a coluna "Suites", e a função, é do tipo `lambda` - funções personalizadas que podem ser escritas em uma única linha.

Nesta função, x equivale a cada linha da coluna de suítes. Então, fizemos uma condicional if em que se o valor de x, ou seja, o registro da linha, for maior que zero, deve ser colocado "Sim", caso contrário (else), coloca-se "Não".

`````python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else "Não")
dados.head()
`````

# 📌 Criando colunas em um DataFrame no Pandas

Existem diversas maneiras de criar colunas em um DataFrame no **pandas**.  
Aqui estão alguns exemplos:

---

## 1️⃣ Atribuição direta de valores a uma nova coluna

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = [7, 8, 9]
df
````

---

## 2️⃣ Criação de uma coluna a partir de operações entre outras colunas

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'] + df['B']
df
```

---

## 3️⃣ Utilização do método `assign()` para criar uma nova coluna

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df = df.assign(C=[7, 8, 9])
df
```

---

## 4️⃣ Utilização do método `apply()` para aplicar uma função e criar uma nova coluna

```python
import pandas as pd

df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df['C'] = df['A'].apply(lambda x: x * 2)
df
```
