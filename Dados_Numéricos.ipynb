### Para saber mais: precisão de valores numéricos

Ao lidar com grandes volumes de dados e pouca memória disponível, é comum utilizar tipos de dados mais compactos. Porém, é essencial garantir que essa escolha não comprometa a precisão dos resultados.

#### Bits e Bytes

* **Bit**: menor unidade de informação digital (0 ou 1).
* **Byte**: conjunto de 8 bits.

#### Tipos de inteiros (int)

* **int64**: utiliza 8 bytes (64 bits), representando números de
  **-9.223.372.036.854.775.808 a 9.223.372.036.854.775.807**.
* **int32**: utiliza 4 bytes (32 bits), representando números de
  **-2.147.483.648 a 2.147.483.647**.
* Outros tipos (mais compactos, mas com limites menores):

| Tipo  | Bits | Valor mínimo               | Valor máximo              |
| ----- | ---- | -------------------------- | ------------------------- |
| int8  | 8    | -128                       | 127                       |
| int16 | 16   | -32.768                    | 32.767                    |
| int32 | 32   | -2.147.483.648             | 2.147.483.647             |
| int64 | 64   | -9.223.372.036.854.775.808 | 9.223.372.036.854.775.807 |

**Dica:** use o menor tipo que seja suficiente para representar seus dados — por exemplo, `int32` para valores pequenos, `int64` para números científicos ou muito grandes.

#### Tipos de ponto flutuante (float)

* **float64**: 64 bits, até **15 dígitos de precisão**.
* **float32**: 32 bits, até **7 dígitos de precisão**.

Assim como nos inteiros, a escolha depende do equilíbrio entre **precisão** e **uso de memória**.

.apply() com lambda, que aplica uma função específica a cada elemento, no caso, a remoção dos caracteres.

###Transformação em múltiplas colunas
####Separar as duas colunas em um dataframe

```python
dados[['taxa_deposito','taxa_limpeza']]
```

`.applymap()`

```python
#dados[['taxa_deposito','taxa_limpeza']].applymap(lambda x: float(x.replace('$', '').replace(',','').strip()))
#Forma antiga, desatualizada
```

>O map em versões anteriores era utilizado apenas em uma Series (coluna única), mas em versões mais recentes do Pandas, o comportamento desse método foi expandido para também funcionar com DataFrames, aplicando a função elemento a elemento em todas as células, de forma semelhante ao que applymap fazia.

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[['taxa_deposito','taxa_limpeza']].map(
    lambda x: float(x.replace('$', '').replace(',','').strip()))
```

###Transformar valores para float

```python
dados[['taxa_deposito','taxa_limpeza']] = dados[
    ['taxa_deposito','taxa_limpeza']].astype(np.float64)
```
