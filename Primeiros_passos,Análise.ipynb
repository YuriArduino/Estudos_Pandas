Método `groupby` do Pandas

O método **`groupby`** do Pandas permite agrupar e resumir dados de um **DataFrame**, com base em um ou mais critérios.  
Esses critérios podem ser variáveis numéricas ou categóricas, como colunas ou índices do DataFrame.

A ideia por trás do `groupby` é **dividir os dados em grupos** com base nos critérios selecionados e, em seguida, aplicar uma operação a esses grupos.  
Essa operação pode ser uma função de agregação, como **soma, média, contagem, desvio padrão**, entre outras, ou mesmo uma operação personalizada definida pela pessoa usuária.

---

### Principais parâmetros do `groupby`

- **`by`**  
  Usado para especificar a coluna ou colunas que queremos agrupar.  
  Como argumento, podemos passar o nome de uma coluna ou uma lista de nomes de colunas.

- **`axis`**  
  Especifica o eixo ao longo do qual queremos agrupar.  
  - Valor padrão: `0` → agrupa pelas **linhas**  
  - Se definido como `1` → agrupa pelas **colunas**

- **`sort`**  
  Define se os grupos devem ser ordenados pelo valor da coluna de agrupamento.  
  - Tipo: booleano (`True` ou `False`)  
  - Valor padrão: `True`

- **`dropna`**  
  Controla se os valores ausentes (`NaN`) serão excluídos durante o processo de agrupamento.  
  - Valor padrão: `True`

# Renomeando colunas em um DataFrame

Em alguns casos, é necessário alterar os **nomes das colunas** em um DataFrame.  
Por exemplo, ao criar o seguinte DataFrame:

```python
df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')
````

O resultado foi:

| Tipo               | Valor       |
| ------------------ | ----------- |
| Studio             | 0.000173078 |
| Loft               | 0.00220674  |
| Casa de Vila       | 0.0107741   |
| Flat               | 0.0205963   |
| Quitinete          | 0.0361733   |
| Casa               | 0.0418415   |
| Casa de Condomínio | 0.0430964   |
| Apartamento        | 0.845139    |

---

Podemos notar que a coluna com os **percentuais** ficou nomeada como `"Tipo"`.
Se quisermos mudar o nome dela para `"Percentuais"`, usamos o método **`.rename()`**:

```python
# Salvando o DataFrame em uma variável
df_exemplo = df['Tipo'].value_counts(normalize=True).to_frame().sort_values('Tipo')

# Renomeando a coluna
df_exemplo.rename(columns={'Tipo': 'Percentuais'}, inplace=True)

# Visualizando o DataFrame
df_exemplo
```

Resultado:

| Percentuais        |             |
| ------------------ | ----------- |
| Studio             | 0.000173078 |
| Loft               | 0.00220674  |
| Casa de Vila       | 0.0107741   |
| Flat               | 0.0205963   |
| Quitinete          | 0.0361733   |
| Casa               | 0.0418415   |
| Casa de Condomínio | 0.0430964   |
| Apartamento        | 0.845139    |

---

✅ O parâmetro **`inplace=True`** garante que a alteração seja aplicada **diretamente** no DataFrame.
Agora sabemos como **renomear colunas** facilmente no Pandas!
