Tratamento de dados nulos no Pandas

O Pandas oferece várias formas de lidar com valores **nulos (`NaN`)** em um DataFrame.  
As principais são:

---

🔹 1. Remover os dados nulos
Utilizando o método **`.dropna()`**, podemos remover linhas ou colunas que contenham valores nulos.

```python
# Remove linhas com valores nulos
df.dropna()

# Remove colunas com valores nulos
df.dropna(axis=1)
````

---

🔹 2. Preencher os dados nulos

Com o método **`.fillna()`**, podemos substituir os valores nulos por um valor específico.

```python
# Preencher valores nulos com 0
df.fillna(0)

# Preencher valores nulos com o valor anterior
df.fillna(method="ffill")

# Preencher valores nulos com o valor seguinte
df.fillna(method="bfill")
```

---

🔹 3. Interpolar os dados nulos

O método **`.interpolate()`** preenche valores nulos com base em valores **interpolados** (estimados a partir dos vizinhos).

```python
# Preencher valores nulos usando interpolação linear
df.interpolate()
```

`Salvando arquivos CSV com diferentes separadores`

Por padrão, o método **`.to_csv()`** do Pandas salva os arquivos separados por **vírgula ( , )**, pois esse é o padrão internacional do formato CSV (*Comma-Separated Values*).  

No entanto, em alguns países (como o Brasil), é comum encontrar arquivos CSV separados por **ponto e vírgula ( ; )**, para evitar confusão com números decimais.

---

Diferença entre `,` e `;`
- **`,` (vírgula)** → padrão internacional do CSV, usado principalmente em países que utilizam **ponto ( . ) como separador decimal**.  
- **`;` (ponto e vírgula)** → alternativa prática em países que utilizam **vírgula ( , ) como separador decimal**, como Brasil, França e Alemanha.

---

✅ Exemplo de uso
```python
# Salvar com separador padrão (vírgula)
df.to_csv('dados.csv', index=False)

# Salvar com separador ponto e vírgula
df.to_csv('dados.csv', index=False, sep=';')
````

---

💡 Curiosidade

* Em inglês/americano:

  * Número decimal → `3.14`
  * CSV → `3.14,100,200`

* Em português/brasileiro:

  * Número decimal → `3,14`
  * CSV → `3,14;100;200`

Isso mostra por que em muitos casos no Brasil usamos **`;`** em vez de **`,`** no CSV.

O Pandas oferece diversos métodos para salvar arquivos em diferentes formatos. Alguns exemplos são:

`to_csv():` salva o DataFrame em um arquivo CSV (Comma-Separated Values);

`to_excel()`: salva o DataFrame em um arquivo Excel;

`to_json()`: salva o DataFrame em um arquivo JSON (JavaScript Object Notation).
